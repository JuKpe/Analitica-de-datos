````python
def top_correlations(df, threshold=0.5, top_n=10):
    """
    Devuelve las correlaciones más fuertes de un dataframe.

    Parámetros:
    df        -- dataframe de Pandas
    threshold -- valor mínimo absoluto de correlación a considerar
    top_n     -- número de pares más fuertes a mostrar

    Retorna:
    DataFrame con las correlaciones ordenadas
    """
    corr_matrix = df.corr()
    corr_pairs = corr_matrix.unstack()
    
    # Eliminar correlaciones de una variable consigo misma
    corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
    
    # Eliminar duplicados (porque la matriz es simétrica)
    corr_pairs = corr_pairs.drop_duplicates()
    
    # Filtrar por threshold
    strong_corr = corr_pairs[abs(corr_pairs) >= threshold]
    
    # Ordenar por valor absoluto
    strong_corr = strong_corr.reindex(strong_corr.abs().sort_values(ascending=False).index)
    
    return strong_corr.head(top_n)
````

````python
resultados = top_correlations(employee_df, threshold=0.6, top_n=10)
print(resultados)
````

## 📌 ¿Hay sobreajuste aquí?

### 🚨 Posibles indicios:

* Si este es el rendimiento **en test/validación** y en entrenamiento tuviste, por ejemplo:

  * Clase 0: F1 = 0.99
  * Clase 1: F1 = 0.95

📌 **Signo de sobreajuste típico**:

* Rendimiento casi perfecto en entrenamiento.
* Caída notable de recall/F1 en la clase minoritaria en validación/test.

### 📌 Pero para confirmarlo, se debería comparar con el mismo reporte en entrenamiento.

## 📌 Recomendación:

Si ves este rendimiento:

* Si en entrenamiento todo está en 0.95 - 0.99 y en test baja mucho en la clase minoritaria, hay sobreajuste.

## Configura el modelo antes de entrenarlo. 
Es como definir las reglas del juego antes de empezar a aprender:

La instrucción:

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])

```


## 📌 Detalle de cada parámetro:

### 📌 `optimizer='Adam'`

Le dice al modelo **qué algoritmo de optimización usar** para actualizar los pesos durante el entrenamiento.

* **Adam** (Adaptive Moment Estimation) es uno de los optimizadores más usados porque combina la eficiencia de **Momentum** y **RMSProp**.
* Controla cómo se ajustan los pesos en base a las derivadas parciales (gradientes) del error respecto a los pesos.

🔍 Podrías personalizarlo así:

```python
from tensorflow.keras.optimizers import Adam
Adam(learning_rate=0.001)
```

---

### 📌 `loss='binary_crossentropy'`

Define la **función de pérdida (coste)** que el modelo tratará de minimizar.

* **Binary Crossentropy** se usa en problemas de clasificación binaria (cuando hay dos clases: 0 o 1).
* Mide la diferencia entre las predicciones del modelo (valores entre 0 y 1) y las etiquetas verdaderas.
* Cuanto más se aleje la predicción del valor real, mayor será la penalización.

👉 Fórmula de Binary Crossentropy:

$$
Loss = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$

---

### 📌 `metrics=['accuracy']`

Especifica **qué métricas quieres calcular y reportar durante el entrenamiento y la evaluación** (pero no influye en el proceso de entrenamiento como la función de pérdida sí lo hace).

* Aquí le estás pidiendo que calcule la **accuracy**: proporción de predicciones correctas sobre el total.

Podrías añadir más:

```python
metrics=['accuracy', 'Precision', 'Recall']
```

---

## 📌 Resumen claro:

| Elemento    | Rol                                                             |
| :---------- | :-------------------------------------------------------------- |
| `optimizer` | Cómo se ajustan los pesos (algoritmo de optimización).          |
| `loss`      | Qué función se minimizará para entrenar el modelo.              |
| `metrics`   | Qué métricas se reportan durante el entrenamiento y evaluación. |

---

## 📌 ¿Por qué es importante?

Antes de llamar a `model.fit()`, el modelo **debe compilarse** para indicarle:

* Cómo aprender (optimización).
* Qué objetivo seguir (función de pérdida).
* Cómo reportar su desempeño (métricas).

Sin este paso, no puedes entrenar el modelo.

````python
model_DL.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy', 'Precision', 'Recall'])
````
