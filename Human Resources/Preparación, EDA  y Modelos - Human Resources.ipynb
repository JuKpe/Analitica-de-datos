````python
def top_correlations(df, threshold=0.5, top_n=10):
    """
    Devuelve las correlaciones ms fuertes de un dataframe.

    Parmetros:
    df        -- dataframe de Pandas
    threshold -- valor mnimo absoluto de correlacin a considerar
    top_n     -- nmero de pares ms fuertes a mostrar

    Retorna:
    DataFrame con las correlaciones ordenadas
    """
    corr_matrix = df.corr()
    corr_pairs = corr_matrix.unstack()
    
    # Eliminar correlaciones de una variable consigo misma
    corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
    
    # Eliminar duplicados (porque la matriz es simtrica)
    corr_pairs = corr_pairs.drop_duplicates()
    
    # Filtrar por threshold
    strong_corr = corr_pairs[abs(corr_pairs) >= threshold]
    
    # Ordenar por valor absoluto
    strong_corr = strong_corr.reindex(strong_corr.abs().sort_values(ascending=False).index)
    
    return strong_corr.head(top_n)
````

````python
resultados = top_correlations(employee_df, threshold=0.6, top_n=10)
print(resultados)
````

## Configura el modelo antes de entrenarlo. 
Es como definir las reglas del juego antes de empezar a aprender:

La instruccin:

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])

```


##  Detalle de cada parmetro:

###  `optimizer='Adam'`

Le dice al modelo **qu algoritmo de optimizacin usar** para actualizar los pesos durante el entrenamiento.

* **Adam** (Adaptive Moment Estimation) es uno de los optimizadores ms usados porque combina la eficiencia de **Momentum** y **RMSProp**.
* Controla cmo se ajustan los pesos en base a las derivadas parciales (gradientes) del error respecto a los pesos.

 Podras personalizarlo as:

```python
from tensorflow.keras.optimizers import Adam
Adam(learning_rate=0.001)
```

---

###  `loss='binary_crossentropy'`

Define la **funcin de prdida (coste)** que el modelo tratar de minimizar.

* **Binary Crossentropy** se usa en problemas de clasificacin binaria (cuando hay dos clases: 0 o 1).
* Mide la diferencia entre las predicciones del modelo (valores entre 0 y 1) y las etiquetas verdaderas.
* Cuanto ms se aleje la prediccin del valor real, mayor ser la penalizacin.

 Frmula de Binary Crossentropy:

$$
Loss = - \frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$

---

###  `metrics=['accuracy']`

Especifica **qu mtricas quieres calcular y reportar durante el entrenamiento y la evaluacin** (pero no influye en el proceso de entrenamiento como la funcin de prdida s lo hace).

* Aqu le ests pidiendo que calcule la **accuracy**: proporcin de predicciones correctas sobre el total.

Podras aadir ms:

```python
metrics=['accuracy', 'Precision', 'Recall']
```

---

##  Resumen claro:

| Elemento    | Rol                                                             |
| :---------- | :-------------------------------------------------------------- |
| `optimizer` | Cmo se ajustan los pesos (algoritmo de optimizacin).          |
| `loss`      | Qu funcin se minimizar para entrenar el modelo.              |
| `metrics`   | Qu mtricas se reportan durante el entrenamiento y evaluacin. |

---

##  Por qu es importante?

Antes de llamar a `model.fit()`, el modelo **debe compilarse** para indicarle:

* Cmo aprender (optimizacin).
* Qu objetivo seguir (funcin de prdida).
* Cmo reportar su desempeo (mtricas).

Sin este paso, no puedes entrenar el modelo.
